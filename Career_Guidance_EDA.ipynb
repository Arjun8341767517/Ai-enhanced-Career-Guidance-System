```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'Student_ID': range(1, 21),
    'Interest': ['AI', 'Web Dev', 'Data Science', 'Cybersecurity', 'AI', 'AI', 'Web Dev', 'Cybersecurity',
                 'Data Science', 'Web Dev', 'AI', 'Cybersecurity', 'Web Dev', 'Data Science', 'AI',
                 'Cybersecurity', 'Web Dev', 'AI', 'Web Dev', 'Data Science'],
    'Skill_Level': [8, 7, 9, 6, 9, 8, 7, 5, 9, 6, 8, 5, 7, 9, 8, 6, 7, 9, 6, 9],
    'Preferred_Language': ['Python', 'JavaScript', 'Python', 'Python', 'Python', 'Python', 'JavaScript', 'Python',
                           'Python', 'JavaScript', 'Python', 'Python', 'JavaScript', 'Python', 'Python',
                           'Python', 'JavaScript', 'Python', 'JavaScript', 'Python'],
    'Career_Suggestion': ['AI Engineer', 'Frontend Dev', 'Data Analyst', 'Security Analyst', 'AI Researcher',
                          'ML Engineer', 'Backend Dev', 'Security Engineer', 'Data Scientist', 'Full Stack Dev',
                          'AI Engineer', 'Security Analyst', 'Frontend Dev', 'Data Analyst', 'AI Researcher',
                          'Cybersecurity Specialist', 'Web Dev', 'AI Engineer', 'Frontend Dev', 'Data Scientist']
}
df = pd.DataFrame(data)

# Visualization 1: Interest Distribution
sns.countplot(data=df, x='Interest')
plt.title('Student Interest Distribution')
plt.show()

# Visualization 2: Skill Level by Interest
sns.boxplot(data=df, x='Interest', y='Skill_Level')
plt.title('Skill Level Distribution by Interest')
plt.show()
```
